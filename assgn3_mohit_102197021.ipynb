```python
import copy
import random
class MyEightPuzzle_misplaced:
    def __init__(self,startState,goalState):
        self.currentState=startState
        self.goal=goalState
        self.emptyIndex=self.emptyTileIndex()
        self.prevstate=None
    def up(self):
        if self.emptyIndex>=6 :
            print("cannot move")
            return False
        else:
            self.prevstate=copy.deepcopy(self)
            self.currentState[self.emptyIndex]=self.currentState[self.emptyIndex+3]
            self.currentState[self.emptyIndex+3]=0
            self.emptyIndex=self.emptyIndex+3
            print("Action :UP")
            return True
    def down(self):
        if self.emptyIndex<=2:
            print("cannot move")
            return False

        else:
            self.prevstate=copy.deepcopy(self)
            self.currentState[self.emptyIndex]=self.currentState[self.emptyIndex-3]
            self.currentState[self.emptyIndex-3]=0

            self.emptyIndex=self.emptyIndex-3
            print("Action :DOWN")
            return True
    def left(self):
        
        if self.emptyIndex==2 or self.emptyIndex==5 or self.emptyIndex==8:
            print('cannot move')
            return False

        else:
            self.prevstate=copy.deepcopy(self)
            self.currentState[self.emptyIndex]=self.currentState[self.emptyIndex+1]
            self.currentState[self.emptyIndex+1]=0
            self.emptyIndex=self.emptyIndex+1
            print("Action :LEFT")
            return True
    def right(self):
        if self.emptyIndex==0 or self.emptyIndex==3 or self.emptyIndex==6:
            print('cannot move')
            return False
        else:
            self.prevstate=copy.deepcopy(self)
            self.currentState[self.emptyIndex]=self.currentState[self.emptyIndex-1]
            self.currentState[self.emptyIndex-1]=0
            self.emptyIndex=self.emptyIndex-1
            print("Action :RIGHT")
            return True
    def emptyTileIndex(self):
        for i in range(0,9):
            if(self.currentState[i]==0):
                return i
    def isGoalReached(self):
        if(self.currentState==self.goal):
            return True
        else:
            return False
    def displayState(self):
        print('-------------------')
        for i in range(0,8,3):
            print(self.currentState[i],self.currentState[i+1],self.currentState[i+2])
        print(f"heuristic:{self.heuristic_misplaced()}")
        print('-------------------')
    def __eq__(self,other):
        return self.currentState==other.currentState
    def find_index(self):
        pass
    def convert_index(self,i):
        x=i//3+1
        y=i%3+1
        ans=[]
        ans.append(x)
        ans.append(y)
        return ans
    def heuristic_euclidean(self):
        euclidean=0
        for i in range(9):
            if self.currentState[i]!=0:
               initial_cord= self.convert_index(i)
               goal_cord=self.convert_index(self.goal.index(self.currentState[i]))
               #print(f"{self.currentState[i]}: current coordinates:{initial_cord}   goal coordinates :{goal_cord} ")
               euclidean=euclidean+(((initial_cord[0]-goal_cord[0])**2)+((initial_cord[1]-goal_cord[1])**2))**0.5
        return euclidean
    def heuristic_manhattan(self):
        manhattan=0
        for i in range(9):
            if self.currentState[i]!=0:
               initial_cord= self.convert_index(i)
               goal_cord=self.convert_index(self.goal.index(self.currentState[i]))
               #print(f"{self.currentState[i]}: current coordinates:{initial_cord}   goal coordinates :{goal_cord}")
               manhattan=manhattan+(abs(initial_cord[0]-goal_cord[0]))+(abs(initial_cord[1]-goal_cord[1]))
        return manhattan
    def heuristic_misplaced(self):
        count=0
        for i in range(9):
            if self.currentState[i]!=0:
                if self.currentState[i]!=self.goal[i]:
                    count+=1
        return count              
    def Possiblenextstates(self):
        stateList=[]
        left_state=copy.deepcopy(self)
        if left_state.left():
            stateList.append(left_state)
        right_state=copy.deepcopy(self)
        if right_state.right():
            stateList.append(right_state)
        up_state=copy.deepcopy(self)
        if up_state.up():
            stateList.append(up_state)
        down_state=copy.deepcopy(self)
        if down_state.down():
            stateList.append(down_state)
        return stateList
    def __gt__(self,other):
        return self.heuristic_misplaced()>other.heuristic_misplaced()
    def __lt__(self,other):
        return self.heuristic_misplaced()<other.heuristic_misplaced()
def best_first(startstate):
    open=[]
    closed=[]
    open.append(startstate)
    while open:
        thisstate=open.pop(0)
        thisstate.displayState()
        if thisstate not in closed:
            closed.append(thisstate)
            if thisstate.isGoalReached():
                print("goal state reached constructing path from goal to start")
                constructpath(thisstate)
                break
            nextStates=thisstate.Possiblenextstates()
            for eachstate in nextStates:
                if eachstate not in closed and eachstate not in open:
                    open.append(eachstate)
            open.sort()
def constructpath(gstate):
    while gstate is not None:
        gstate.displayState()
        gstate=gstate.prevstate

```

```python
import random
def hill_climbing(startstate):
    open=[]
    goal=0
    open.append(startstate)
    while open:
        thisstate=open.pop(0)
        thisstate.displayState()
        if thisstate.isGoalReached():
            print("goal state reached constructing path from goal to start")
            goal=1
            constructpath(thisstate)
            break
        nextStates=thisstate.Possiblenextstates()
        while nextStates:
            move=random.randint(0,len(nextStates)-1)                                  
            print(f"{move} heuristic={nextStates[move].heuristic_misplaced()}    this heuristic={thisstate.heuristic_misplaced()}")             
            if nextStates[move].heuristic_misplaced()<thisstate.heuristic_misplaced():
                open.append(nextStates[move])
                print(f"appending ")
                nextStates[move].displayState()
                break
            else:
                nextStates.pop(move)
    if goal==0:
        print("hill climbing method failed")
def constructpath(gstate):
    while gstate is not None:
        gstate.displayState()
        gstate=gstate.prevstate
```

```python
import math
import copy
class TSP():
    def __init__(self,thismap,startcity):
        TSP.map=thismap
        self.startcity=startcity
        self.currentcity=startcity
        self.cost=0
        self.visitedlist=[]
        self.visitedlist.append(self.currentcity)
    def displaystate(self):
        print("***********")
        print(f"current city:{self.currentcity}    visitited cities={self.visitedlist}   cost={self.cost}")
        print("***********")
    def __gt__(self,other):
        return self.cost > other.cost
    def __lt__(self,other):
        return self.cost < other.cost
    def __eq__(self,other):
        return self.visitedlist==other.visitedlist
    def goalachieved(self):
        if self.currentcity==4:
            return True
        else:
            return False
    def move(self,city):
        if city!=self.currentcity and city not in self.visitedlist:
            print(f"moving from city {self.currentcity} to {city}")
            self.cost+=TSP.map[self.currentcity][city]
            self.currentcity=city
            self.visitedlist.append(self.currentcity)
            return True
        elif len(self.visitedlist)==len(TSP.map[0]):
            print(f"moving from city {self.currentcity} to {self.startcity}")
            self.cost+=TSP.map[self.currentcity][self.startcity]
            self.currentcity=self.startcity
            self.visitedlist.append(self.startcity)
            return True
        else:
            print("already visited")
            return False
    def possiblenextstates(self):
        statelist=[]
        for i in range(0,len(TSP.map[0])):
            state=copy.deepcopy(self)
            if state.move(i):
                statelist.append(state)
        return statelist
    
def UCS(state):
    open=[]
    closed=[]
    open.append(state)    
    while open:
        thisstate=open.pop(0)
        thisstate.displaystate()
        if thisstate not in  closed:
            closed.append(thisstate)
            if thisstate.goalachieved():
                print("goal state found .. ")
                thisstate.displaystate()
                break
            nextstate=thisstate.possiblenextstates()
            for eachstate in nextstate:
                if eachstate not in closed and eachstate not in open:
                    open.append(eachstate)
                    open.sort()

map=[[0,3,1,math.inf,math.inf,math.inf],
     [math.inf,0,math.inf,3,math.inf,math.inf],
     [math.inf,math.inf,0,1,2,math.inf],
     [math.inf,math.inf,math.inf,0,3,math.inf],
     [math.inf,math.inf,math.inf,math.inf,0,math.inf],
     [math.inf,math.inf,math.inf,math.inf,math.inf,0]]
#A=0
#B=1
#C=2
#D=3
#G=4
#S=5
start=0
problem=TSP(map,start)
UCS(problem)
```
